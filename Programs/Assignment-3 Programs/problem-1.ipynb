```python
## now for generative linear regression 
#solve ERM using GD ???
#in generative linear regression using MLE the sample mean and the sample deviation are the ones to go for !!
m = y_data.shape[0]
d_y = y_data.shape[1]
mu_2 = np.sum(y_data, axis = 0)/(y_data.shape[0])
print("means of labels = ", mu_2)
sigma_2  = np.zeros((d_y,d_y))
for i in range(d_y):
    for j in range (d_y):
        sigma_2[i][j] = (1/m)* np.sum((y_data[:, i] - mu_2[i])*(y_data[:, j] - mu_2[j]), axis = 0)

print("Covaraince matrix = ", sigma_2)
```

```python
# now to do the same thing but with the joint
#z = [x y]
z = data.iloc[1:,:]
z = z.to_numpy()
m = z.shape[0]
d = z.shape[1]
mean = np.mean(z, axis  = 0)
print(mean.shape)
cov  = np.ones((d, d))
for i in range (d):
    for j in range (d):
        cov[i][j]  = np.mean((z [:, i] - mean[i]) * (z[:, j] - mean[j]), axis = 0)

print(cov.shape)
precision = np.linalg.inv(cov)
A_11 = precision[0:107, 0:107]
A_22 = precision[107:109 , 107:109]
A_12 = precision[0:107 , 107:109]
A_21 = precision[107:109, 0:107]
mu_1 = mean[0:107]
mu_2 = mean[107:109]
bayes_coeff = np.linalg.inv(A_22)@A_21
bayes_constant  = mu_2 - bayes_coeff @ mu_1 
print(bayes_coeff.shape)
def bayes_predictor(bayes_coef, X_test, bayes_constant):
    return X_test @ bayes_coef.T + bayes_constant

y_pred = bayes_predictor(bayes_coeff, X_train, bayes_constant)
metrics.explained_variance_score(y_train, y_pred)
```
