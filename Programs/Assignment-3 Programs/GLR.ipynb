```python
class Model:
    def __init__(self, function,data):
        self.func = function
        self.data = data
    def predict(self, X):
        matrix = np.apply_along_axis(self.func, arr = X, axis=1, data= self.data)
        return matrix
    def explained_accuracy(self, X, y):
        return metrics.explained_variance_score(y, self.predict(X), multioutput='raw_values')
```

```python
def GLG_pred(X, data):
    precision = data['precision']
    y_size = data['y_size']
    mu_x = data['mu_x']
    mu_y = data['mu_y']
    A_11 = precision[:-y_size,:-y_size]
    A_22 = precision[-y_size:,-y_size:]
    A_12 = precision[:-y_size, -y_size:]
    A_21 = precision[-y_size:, :-y_size]
    return (-1*np.linalg.inv(A_22) @ A_21 @ (X  - mu_x)) + mu_y
```
