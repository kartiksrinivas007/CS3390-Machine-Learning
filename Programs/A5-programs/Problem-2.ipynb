```python
import scipy as sc
import numpy as np
from sklearn import * 
import pandas as pd

```

```python
data = sc.io.loadmat('Datasets/DrivFace/DrivFace.mat')
data = data
print(data.keys())
print(data['drivFaceD'][0][0][0].shape)
X_data = data['drivFaceD'][0][0][0]
X_dataframe = pd.DataFrame(X_data)
display(X_dataframe)
```

```python
label_data = pd.read_csv('Datasets/DrivFace/drivPoints.txt')
xF = label_data['xF']
yF = label_data['yF']
wF = label_data['wF']
hF = label_data['hF']
y_data = label_data[["xF", "yF", "wF", "hF"]]
display(y_data)
y_data = y_data.to_numpy()
```

```python
X_data.shape
y_data.shape
```

```python
X_train,X_test, y_train, y_test = model_selection.train_test_split(X_data, y_data, test_size= 0.5 ,random_state=42, shuffle = True)
print(X_train.shape)
print(y_train.shape)
```

```python
alpha_vals = [0.01, 0.1, 1, 10, 100]
#setup ridge regression
best_score = 0
best = [0, 0]
best_ridge = linear_model.Ridge()
best_a = 0
a_list_inc = []
for a in alpha_vals:
    model_ridge = linear_model.Ridge(alpha = a)
    scores = model_selection.cross_val_score(model_ridge, X_train, y_train)
    print(scores.mean())
    if(scores.mean() > best_score):
        best_score = scores.mean()
        best_ridge = model_ridge
        best_a = a
        a_list_inc.append(a)
print(best_ridge)
best_ridge.fit(X_train, y_train)
print(best_ridge.score(X_test, y_test))
print(a_list_inc) # continuous improvement of the scores
```

```python
model_unreg = linear_model.LinearRegression(fit_intercept=True)
model_unreg.fit(X_train, y_train)
print("Unregularized score = ", model_unreg.score(X_test, y_test))
```
