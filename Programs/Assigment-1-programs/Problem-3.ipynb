```python
import numpy as np
import pandas as pd
from sklearn import *
from sklearn.model_selection import *
```

```python
feature_data = pd.read_excel('Datasets/LSVT_voice_rehabilitation/LSVT_voice_rehabilitation.xlsx', 0)
label_data = pd.read_excel('Datasets/LSVT_voice_rehabilitation/LSVT_voice_rehabilitation.xlsx', 1)
```

```python
label_data.loc[label_data["Binary class 1=acceptable, 2=unacceptable"] == 1] = 0
label_data.loc[label_data["Binary class 1=acceptable, 2=unacceptable"] == 2] = 1
print(label_data)
```

```python
Net_data = pd.concat([feature_data,label_data], axis = 1)
display(Net_data)
```

```python
d_train, d_test = train_test_split(Net_data, test_size =0.5, shuffle=True, random_state=42)
print(len(d_train))
print(type(d_train))
```

```python
X_train = d_train.iloc[:, :-1]
X_test = d_test.iloc[:, :-1] 
print(type(X_train))
print(type(X_test))
display(X_train)
```

```python
y_train = d_train.iloc[:,-1:]
y_test = d_test.iloc[:,-1:]
```

```python
print(y_train)
print(y_test)
```

```python
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
X_train = X_train.to_numpy() #rerunning might cause issue because of changed type in cached memory
X_test = X_test.to_numpy()
```

```python
model = linear_model.Perceptron(max_iter = 10000)
print(y_train.ravel())
model.fit(X_train, y_train.ravel()) #needs 1D array shape
```

```python
train_score = model.score(X_train, y_train)
test_score = model.score(X_test,y_test)
```

```python
print("Score using Perceptron and identity map = ", test_score)
```

```python
#now to apply a polynomial kernel
X_new_train = np.concatenate((X_train, np.square(X_train)), axis = 1)
X_new_test = np.concatenate((X_train, np.square(X_train)), axis = 1)
model.fit(X_new_train, y_train.ravel())
y_pred = model.predict(X_new_test) 
polynomial_kernel_score = model.score(X_new_test, y_test.ravel())
print("Score using perceptron and polynomial kernel = ", polynomial_kernel_score)
```

```python
logreg = linear_model.LogisticRegression(fit_intercept=True)
logreg.fit(X_new_train, y_train.ravel())
logreg.score(X_new_test, y_test.ravel())
#proves non convergence of logistic regression
```

```python
logreg_unbiased = linear_model.LogisticRegression(fit_intercept=False)
logreg_unbiased.fit(X_train, y_train.ravel())
print("score using unbiased identity feature map and Logistic regression = " , logreg_unbiased.score(X_test, y_test.ravel()))
```
